```python
m1 = Matrix([
    [1, 2],
    [3, 4]
])

m2 = Matrix([
    [2, 5],
    [6, 1]
])

print("The determinanat of this matrix:")
print(m1)
print("is", m1.determinant())
print('\n')

print("The determinanat of this matrix:")
print(m2)
print("is", m2.determinant())
print('\n')

print("The trace of this matrix:")
print(m1)
print("is", m1.trace())
print('\n')

print("The trace of this matrix:")
print(m2)
print("is", m2.trace())
print('\n')

print("The inverse of this matrix:")
print(m1)
print("is", m1.inverse())
print('\n')

print("The inverse of this matrix:")
print(m2)
print("is", m2.inverse())
print('\n')

print("The transpose of this matrix:")
print(m1)
print("is", m1.T())
print('\n')

print("The transpose of this matrix:")
print(m2)
print("is", m2.T())
print('\n')

m3 = m1 + m2
print("m1 + m2 is")
print(m3)
print('\n')

m3 = m1 - m2
print("m1 - m2 is")
print(m3)
print('\n')

negativeM1 = -m1
print("Negative m1 is")
print(negativeM1)
print('\n')

negativeM2 = -m2
print("Negative m2 is")
print(negativeM2)
print('\n')

product = m1 * m2
print("m1 * m2 is")
print(product)
print('\n')

doubled = 2 * m1
print("2 * m1 is")
print(doubled)
```
